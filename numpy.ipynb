```python
import matplotlib.pyplot as plt

x = np.array([0,4])
y =np.array([0,4])

plt.plot(x,y)
plt.show()
```

```python
x =np.array([0,1,2,3,4])
y =np.array([0,1,2,3,4])

plt.plot(x,y, marker = "*")
plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
x_axis =np.array([0,3,5,7])
y_axis=np.array([0,5,8,16])

plt.scatter(x_axis,y_axis)
plt.show
```

```python
x =np.array(["Apples","Banana","Mango","Grapes"])
y =np.array([4,15,20,2])

plt.bar(x,y,color="black",width=0.8)
plt.show()
```

```python
x =np.array(["Apples","Banana","Mango","Grapes"])
y =np.array([4,15,20,2])

plt.barh(x,y,height =0.9)
plt.show()
```

```python
x = np.array([70,50,10,90])
my_label =["BMW","Jeep","Toyota","Volvo"]
my_colors =["green","red","black","blue"]

my_explode =[0.3,0,0,0]
plt.pie(x,labels=my_label,colors=my_colors, explode =my_explode, shadow= True)


plt.title("cars")
plt.legend(title="cars")
plt.show()
```
